```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
# load image
img = cv.imread('kanye.jpg')
```

```python
# plot the image
plt.imshow(img)
```

```python
# convert from BGR to RGB (for plotting)
convert = lambda x: cv.cvtColor(x, cv.COLOR_BGR2RGB)
RGB_img = convert(img)
```

```python
plt.imshow(RGB_img)
```

```python
# Look at shape of picture
img.shape
```

```python
# split in channels

b,g,r = cv.split(img)
plt.imshow(g)
```

```python
# merge the channels
img = cv.merge((b,g,r))
```

```python
# kill red channel
img[:,:,0] = 0
RGB_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

```python
# scale image

scale_percent = 10 # percent of original size
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)
# resize image
resized = cv.resize(img, dim, interpolation = cv.INTER_AREA) 
```

```python
RGB_img = convert(resized)
plt.imshow(RGB_img)
```

```python
# select pixel patch to modify
RGB_img[0:10,0:10]=0
plt.imshow(RGB_img)
```

```python
# select pixel patch to modify
RGB_img[150:170,150:170]=255
plt.imshow(RGB_img)
```

```python
# apply some filter
edges = cv.Canny(RGB_img,100,200)
plt.imshow(edges)
```

```python
import numpy as np
kernel = np.ones((3,3),np.float32)/9
dst = cv.filter2D(RGB_img,-1,kernel)
dst.shape
plt.imshow(dst)
print(kernel)
```

```python
kernel2 = np.ones((3,3),np.float32)/9
print(kernel2)
kernel2[0,:]=0
print(kernel2)
```
